* **How would you define Machine Learning?**

It is a sub-component of the Artificial Intelligence. Basically, It offer us an algorithm that predict output via training the dataset given beforehand.

---
* **What are the differences between Supervised and Unsupervised Learning? Specify example 3 algorithms for each of these.**

In supervised learning, we give the machine both input (features) and outputs (labels) as a dataset and then we expect from machine to offer us an algorithm via train-test-split method. "Regression" and "Classification" are two types of supervised machine learning techniques. However, in unsupervised learning, we give the machine only inputs and then we expect from the machine offer us an algorithm. "Clustering" and "Association" are two types of unsupervised learning.
  1. Examples of supervised learning: Predict the expected lifespan via blood pressure, blood cholesterol, blood glucose, waist circumferences, serum ALT, serum AST values of a person.
  2. Examples of unsupervised learning: To forecast a possible similarity with a given disease via some molecular markers.
---
* **What are the test and validation set, and why would you want to use them?**

Due to establish an algorithm, the machine is required to a training set. For this reason 30% of the whole dataset is used by machine. Then, machine testing the power of algorithm which already establish via rest of the data set. Once an algorithm has been created, given a new data, machine divide the data into training, validation and testing sets. Then, it checks the algorithm's validity by validation. Also there is one more method that named "Cross Validatin". This method conduct the train-validity-test process at once.

---
* **What are the main preprocessing steps? Explain them in detail. Why we need to prepare our data?**

Machine Learning is often using complicated and large datasets. By nature, this kind of datasets includes very vary data. However, to obtain a useful algorithm and avoid errors, we must use homogenous and appropriate data.
Therefore, before using the dataset, we should explore the dataset and fix the problems. (Exploratory Data Analysis(EDA)) For this purpose, there are seven main steps:
1. Duplicate values: Dataset should investigate and than, it should be clear the duplicate values to avoid.
2. Imbalanced data: Some dataset involve more than one classes. And sometimes variable of one class of this dataset should be extremely more or less than other. In this case, we should use "oversampling" or "undersampling" method to fix this distortion.
3. Missy value: Sometimes a dataset may include many null index. To fix this problem we can use either "Eleminate missing values" or "Filling with mean or median" method. For this last method, if our values is normally distribute we fill the null index with mean of the values. if our values is not normally distribute we fill the null index with median of the values.
4. Determine the outlier: For clear the marginal values this method should be used.
5. Feature scaling: It can be use for normalisation or standardization the our values
6. Feature extraction: It is to obtain useful, calculable information from the information in the dataset.
7. Feature encoding: It is a method that used for transforming the nominal or ordinal data to able to use in the machine learning process.

#### 10.4. $R^2$ score and comparasion (Model score, $R^2$ or coeff of determinant)


Adjusted $R^2$ also indicates how well terms fit a curve or line, but adjusts for the number of terms in a model. If you add more and more useless variables to a model, adjusted r-squared will decrease. If you add more useful variables, adjusted r-squared will increase. Adjusted $R^2$ will always be less than or equal to $R^2$.

```python
# You always use this algorithm for calculate the R^2 score
def adj_r2 (X,y,model):
    """
    X: input
    y: output
    model: regression model
    """
    r_squared = model.score(X,y)
    return(1 - (1-r_squared)*(len(y)-1)/(len(y)-X.shape[1]-1))
```

```python
#R^2 = 1-(RSS/TSS) = Regression error/TSS 
# R^2 Lineer Regression
print("Adj. R2 of the train set",adj_r2(X_train_2,y_train_2,model_2))
print("Adj. R2 of the test set",adj_r2(X_test_2,y_test_2,model_2))
#Lasso
print("Lasso Train: ", model_lasso.score(X_train_2, y_train_2))
print("Lasso Test: ", model_lasso.score(X_test_2, y_test_2))
#R^2 Lasso
print("Adj. R2 of the lasso train set",adj_r2(X_train_2,y_train_2,model_lasso))
print("Adj. R2 of the lasso test set",adj_r2(X_test_2,y_test_2,model_lasso))
```
